```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
import zipfile
```

```python
train_df = pd.read_csv("/content/train.csv.zip" , compression = "zip")
test_df = pd.read_csv("/content/test.csv.zip" , compression = "zip")

train_df
```

```python
y_train = train_df["label"]
X_train = train_df.drop(labels = ["label"] , axis = 1)

```

```python
#Normalization...
X_train = X_train/255.00
test_df = test_df/255.00


#reshape
X_train = X_train.values.reshape(len(X_train) , 28 , 28 , 1)
test_df = test_df.values.reshape(len(test_df) , 28 , 28 , 1)
y_train = tf.one_hot(y_train , 10)
```

```python
import numpy as np
import keras
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Dense, Input
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten
from keras import backend as k
```

```python
inpx = Input(shape=(28,28,1))
layer1 = Conv2D(32, kernel_size=(3, 3), activation='relu')(inpx)
layer2 = Conv2D(128, (3, 3), activation='relu')(layer1)
layer3 = MaxPooling2D(pool_size=(3, 3))(layer2)
layer4 = Dropout(0.5)(layer3)
layer5 = Flatten()(layer4)
layer6 = Dense(512, activation='relu')(layer5)
layer7 = Dense(10, activation='softmax')(layer6)
```

```python
model = Model([inpx], layer7)
model.compile(optimizer=keras.optimizers.Adadelta(),
              loss=keras.losses.categorical_crossentropy,
              metrics=['accuracy'])
#model.summary()
model.fit(X_train, y_train, epochs=1, batch_size=32)
```

```python
L = model.predict(test_df)

L
```

```python
prediction = np.zeros(len(L))
for i in range(len(L)):
    value = L[i][0]
    index = 0
    for j in range(len(L[0])):
        if L[i][j] > value:
            value = L[i][j]
            index = j
    prediction[i] = index

prediction
```

```python
actual_test_df = pd.read_csv("/content/test.csv.zip" , compression = "zip")
actual_test_df["label"] = prediction

convert_dict = {"label" : int}
actual_test_df = actual_test_df.astype(convert_dict) #converts final predictions to int type
# this are the final predictions.......
```
